```python
def get_headers():
    return {
        'User-Agent': 'PostmanRuntime/7.24.1',
        'Accept': '*/*',
        'Connection': 'keep-alive'
    }
```

```python
def get_params(regNumber):
    return {'regNumber': str(regNumber)}
```

```python
def get_info(regNumber):
    _url_common_info = 'https://zakupki.gov.ru/epz/order/notice/ea44/view/common-info.html'
    response = requests.get(url=_url_common_info, headers=get_headers(), params=get_params(regNumber=regNumber))
    return response
```

```python
notificationNumber = '0146200003220000016'
```

```python
z = get_info(notificationNumber)
```

```python
def get_blocks(response):
    soup = BeautifulSoup(response.content, 'html.parser')
    return soup.find_all('div', class_='row blockInfo')
```

```python
blocks = get_blocks(z)
```

```python
def get_block_with(blocks, text: str):
    for block in blocks:
        if text in block.getText():
            return block
    return None
```
